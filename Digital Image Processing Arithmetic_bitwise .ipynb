```python
from matplotlib import pyplot as plt

import cv2
import numpy as np

image1 = cv2.imread('/content/drive/MyDrive/DIgital Image Processing/bit1.png')
image2 = cv2.imread('/content/drive/MyDrive/DIgital Image Processing/bit2.png')

plt.imshow(image1)
print("Original Image 1")
plt.show()
plt.imshow(image2)
print("Original Image 2")
plt.show()

# cv2.bitwise_and is applied over the
# image inputs with applied parameters
dest_and = cv2.bitwise_and(image2, image1, mask = None)

# the window showing output image
# with the Bitwise AND operation
# on the input images
plt.imshow(dest_and)
print("Bitwise And: ")
plt.show()

# De-allocate any associated memory usage
if cv2.waitKey(0) & 0xff == 27:
	cv2.destroyAllWindows()

```
